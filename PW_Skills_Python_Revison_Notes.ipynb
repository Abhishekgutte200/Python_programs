An **identifier** is simply the name used to identify a variable, function, class, module, or other objects in Python. It follows certain rules:

1. It must start with a letter (A-Z or a-z) or an underscore (_).
2. The rest of the name can include letters, digits (0-9), or underscores.
3. Identifiers are case-sensitive, meaning myVar and myvar would be two different identifiers.
4. Reserved keywords (like if, for, class, etc.) cannot be used as identifiers.

**Examples of valid identifiers:**

1. my_variable
2. userName
3. _count

**Examples of invalid identifiers:**

1. 2variable (cannot start with a number)
2. if (reserved keyword)

In Python, predefined keywords (also known as reserved words) are special words that have a specific meaning to the Python interpreter. These keywords are part of the Python syntax and cannot be used as variable names, function names, or identifiers because they are reserved for specific purposes in the language.

Some examples of Python's predefined keywords include:

1. if, else, elif: For conditional statements.
2. for, while: For loops.
3. def: To define a function.
4. class: To define a class.
5. return: To return a value from a function.
6. try, except: For handling exceptions.
7. import: To import a module.
8. True, False: Boolean values.
9. None: Represents the absence of a value.

Sometimes we need to declare a function without any name. The nameless property function is called an anonymous function or lambda function.

```python
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8
```

```python
greet = lambda: "Hello, World!"
print(greet())  # Output: "Hello, World!"
```

```python
# Program for even numbers without lambda function
def even_numbers(nums):
    even_list = []
    for n in nums:
        if n % 2 == 0:
            even_list.append(n)
    return even_list

num_list = [10, 5, 12, 78, 6, 1, 7, 9]
ans = even_numbers(num_list)
print("Even numbers are:", ans)
```

```python
# Program for even number with a lambda function
l = [10, 5, 12, 78, 6, 1, 7, 9]
even_nos = list(filter(lambda x: x % 2 == 0, l))
print("Even numbers are: ", even_nos)
```

**Map**

image.png

In Python, the map() function is used to apply some functionality for every element present in the given sequence and generate a new series with a required modification.

```python
list1 = [2, 3, 4, 8, 9]
list2 = list(map(lambda x: x*x*x, list1))
print("Cube values are:", list2)
```

**Reduce**

image.png

1. In Python, the reduce() function is used to minimize sequence elements into a single value by applying the specified condition.

2. The reduce() function is present in the functools module; hence, we need to import it using the import statement before using it.

```python
from functools import reduce
list1 = [20, 13, 4, 8, 9]
add = reduce(lambda x, y: x*y, list1)
print("Addition of all list elements is : ", add)
```

**Generators**

A Generator in Python is a function that returns an iterator using the Yield keyword.

image.png

```python
def simpleGeneratorFun():
    yield 1
    yield 2
    yield 3

# Driver code to check above generator function
for value in simpleGeneratorFun():
    print(value)
```

yield: This is the key difference between generators and regular functions. The yield statement pauses the function saving all its states and later continues from there on successive calls.

```python
def square_numbers(numbers):
    for num in numbers:
        yield num * num

# Using the generator
nums = [1, 2, 3, 4, 5]
squares = square_numbers(nums)
'''
for square in squares:
    print(square)
    '''
```
