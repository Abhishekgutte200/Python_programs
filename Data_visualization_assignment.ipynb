# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n_samples = 1000
n_features = 5

corr_matrix = np.array([
    [1.0, 0.8, 0.7, 0.4, 0.2],
    [0.8, 1.0, 0.9, 0.6, 0.3],
    [0.7, 0.9, 1.0, 0.8, 0.5],
    [0.4, 0.6, 0.8, 1.0, 0.7],
    [0.2, 0.3, 0.5, 0.7, 1.0]
])

data = np.random.multivariate_normal(mean=np.zeros(n_features), cov=corr_matrix, size=n_samples)

df = pd.DataFrame(data, columns=[f'Feature {i+1}' for i in range(n_features)])

corr = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)

plt.title('Correlation Matrix')

plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)
df_pivot = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='sum')

fig = go.Figure(data=[go.Heatmap(
    z=df_pivot.values,
    x=df_pivot.columns,
    y=df_pivot.index,
    colorscale='Viridis',
    reversescale=True,
    hoverongaps=False
)])

fig.update_layout(
    title='Variation in Sales Across Months and Days',
    xaxis_title='Day',
    yaxis_title='Month'
)

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(
    x=x,
    y=y,
    z=z,
    colorscale='Viridis',
    reversescale=True,
    hoverinfo='z'
)])

fig.update_layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene = dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'),
    width=800,
    height=600
)

fig.show()
```

# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show

x = np.linspace(0, 10, 100)

y = np.sin(x)

p = figure(title="Sine Wave",
           x_axis_label='x',
           y_axis_label='sin(x)')

p.line(x, y, line_width=2)

show(p)

```

```python
import numpy as np
from bokeh.plotting import figure, show

np.random.seed(0)
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency',
           x_range=[edges[0], edges[-1]])

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white")

show(p)
```
